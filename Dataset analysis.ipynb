```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_parquet('train_data.parquet')
train.head(5)
```

```python
train.shape
```

```python
train.info()
```

```python
train.isnull().sum() # no missing values
```

```python
train.drop(columns = ['timestamp', 'expiry'], inplace = True)
```

```python
train.head()
```

```python
call = []
for col in train.columns.tolist() :
    if 'call' in col:
        call.append(col)
    else:
        continue
len(call)
```

```python
put = []
for col in train.columns.tolist() :
    if 'put' in col:
        put.append(col)
    else:
        continue
len(put)        
```

```python
call_train = train.drop(columns = put)
call_train.shape
```

```python
put_train = train.drop(columns = call)
put_train.shape
```

```python
k_call = []
for col in call:
    col.split('iv')[1]
    k_call.append(col.split('iv')[1].split('_')[1])
len(k_call)   
```

```python
iv = 'call_iv_23500'.split('_')
iv.pop(0)
iv
```

```python
all_values = []
modified = call_train[call]
for i in range(len(modified)):
    value = modified.iloc[i].tolist()  # 26 values
    all_values.extend(value)
len(all_values)
```

```python
base = call_train.drop(columns = call)
base.shape
```

```python
repeated_indices = np.repeat(base.index.values, 26)
expanded_col = base.loc[repeated_indices].reset_index(drop=True)
```

```python
expanded_col.head(26)
expanded_col.shape
```

```python
expanded_col['k'] = k_call * len(base)
```

```python
expanded_col.head(27)
```

```python
expanded_col['iv'] = all_values
```

```python
expanded_col.info()
```

```python
expanded_col['k'] = expanded_col['k'].astype(float)
```

```python
expanded_col.info()
```

```python
expanded_col.to_csv('call_train.csv', index=False)
```

```python
expanded_col.to_parquet('call_train', index = False)
```

```python
iv_values = []
modified = put_train[put]
for i in range(len(modified)):
    value = modified.iloc[i].tolist()  # 26 values
    iv_values.extend(value)
len(iv_values)
```

```python
k_put = []
for col in put:
    col.split('iv')[1]
    k_put.append(col.split('iv')[1].split('_')[1])
len(k_put) 
```

```python
repeated_indices = np.repeat(base.index.values, 26)
expanded_put = base.loc[repeated_indices].reset_index(drop=True)
expanded_put.shape
```

```python
expanded_put['k'] = k_put * len(base)
```

```python
expanded_put['iv'] = iv_values
```

```python
expanded_put.head(27)
```

```python
expanded_put['k'] = expanded_put['k'].astype(float)
```

```python
expanded_put.info()
```

```python
expanded_put.to_parquet('put_train', index = False)
```
